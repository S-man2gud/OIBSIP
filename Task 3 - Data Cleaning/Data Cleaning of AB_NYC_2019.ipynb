# Data Cleaning of AB_NYC_2019 Dataset Project
## Import neccessary libraries

```python
# Import libraries
import pandas as pd
```

### Load dataset

```python
file_path = 'AB_NYC_2019.csv'
data = pd.read_csv(file_path)
```

## Explore the dataset

```python
# Preview the dataset
data.head()
```

```python
# Preview the dataset
data.info()
```

# Handling outliers
### Check for outliers

```python
# Check for outliers
data.isnull().any()
```

## Handling missing values
### Check the null values in the name column

```python
# Check the null values in the name column
data[data['name'].isna()==1]
```

### Fill missing 'name' and 'host_name' with "Unknown"

```python
# Fill missing 'name' and 'host_name' with "Unknown"
data['name'] = data['name'].fillna('Unknown')
data['host_name'] = data['host_name'].fillna('Unknown')
```

### Check if change is effected

```python
# Check if change is effected
data.head(2855)
```

### Fill missing 'reviews_per_month' with 0

```python
# Fill missing 'reviews_per_month' with 0
data['reviews_per_month'] = data['reviews_per_month'].fillna(0)
data.head()
```

### Convert 'last_review' to datetime format, leaving missing values as NaT

```python
# Convert 'last_review' to datetime format, leaving missing values as NaT
data['last_review'] = pd.to_datetime(data['last_review'], errors='coerce')
data.head()
```

## Handling Duplicates
### Check for duplicate rows

```python
# Check for duplicate rows
data.duplicated()
```

### Verify duplicates

```python
# Verify duplicates
data[data['host_id'].duplicated()==1]
```

### Remove duplicate rows if any

```python
# Remove duplicate rows if any
data = data.drop_duplicates()
```

## Handling outliers
### Check for outliers

```python
# Check for outliers in price column
Q1 = data['price'].quantile(0.25)
Q3 = data['price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = data[(data['price'] < lower_bound) | (data['price'] > upper_bound)]
outliers.head()
```

```python
# Check for outliers in minimum_nights column
Q1 = data['minimum_nights'].quantile(0.25)
Q3 = data['minimum_nights'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = data[(data['minimum_nights'] < lower_bound) | (data['minimum_nights'] > upper_bound)]
outliers.head()
```

### Remove outliers

```python
# Remove outliers
data = data[(data['price'] > lower_bound) & (data['price'] < upper_bound)]
data.describe()
```

```python
data = data[(data['minimum_nights'] > lower_bound) & (data['minimum_nights'] < upper_bound)]
data.describe()
```

## Standardize Text in Categorical Columns
### Convert text columns to lowercase and strip whitespace

```python
# Convert text columns to lowercase and strip whitespace
text_columns = ['name', 'host_name', 'neighbourhood_group', 'neighbourhood', 'room_type']
for col in text_columns:
    data[col] = data[col].str.strip().str.lower()
```

## Remove Unnecessary Columns

```python
# Remove Unnecessary Columns
# Drop the 'id' column if not needed for analysis
data = data.drop(columns=['id'])
```

## Final Summary of Cleaned Data

```python
# Final Summary of Cleaned Data
print("Shape of the cleaned dataset:", data.shape)
print("Missing values in cleaned data:\n", data.isnull().sum())
print("Sample of cleaned data:\n", data.head())
```
