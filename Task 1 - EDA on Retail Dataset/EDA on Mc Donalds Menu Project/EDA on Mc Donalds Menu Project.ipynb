# Exploratory Data Analysis on Mc Donalds Menu Project
## Importing neccessary libraries
### Loading Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'menu.csv'

data=pd.read_csv(file_path)
```

## Explore the Data
### Display the first few rows

```python
# Display the first few rows
data.head()
```

### Check data types

```python
# Check data types
data_types = data.dtypes
data_types
```

## Clean Data
### Handle missing values

```python
# Check for missing values
missing_values = data.isnull().sum()
print("Missing Values:\n"), missing_values
```

### Drop rows with missing values

```python
# Drop rows with missing values
data_cleaned = data.dropna()
data_cleaned
```

## Handle duplicates
### Check for duplicates

```python
# Check for duplicates
duplicates = data.duplicated().sum()
print("Duplicate Rows: ", duplicates)
```

## Descriptive Statistics

```python
# Descriptive Statistics
descriptive_stats = data.describe()
print("Descriptive Statistics:\n", descriptive_stats)
```

## Visualization Setup
### Histogram of calories

```python
# Histogram of calories
plt.figure(figsize=(8, 6))
sns.histplot(data['Calories'], bins=20, kde=True)
plt.title('Distribution of Calories')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()
```

### Boxplot of sodium content by category

```python
# Boxplot of sodium content by category
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Sodium', data=data)
plt.xticks(rotation=45)
plt.title('Sodium Content by Food Category')
plt.ylabel('Sodium (mg)')
plt.show()
```

### Boxplot of carbohydrate content by category

```python
# Boxplot of carbohydrate content by category
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Carbohydrates', data=data)
plt.xticks(rotation=45)
plt.title('Carbohydrate Content by Food Category')
plt.ylabel('Carbohydrate (mg)')
plt.show()
```

### Boxplot of protein content by category

```python
# Boxplot of protein content by category
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Protein', data=data)
plt.xticks(rotation=45)
plt.title('Protein Content by Food Category')
plt.ylabel('Protein (mg)')
plt.show()
```

### Boxplot of Dietary Fiber content by category

```python
# Boxplot of Dietary Fiber content by category
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Dietary Fiber', data=data)
plt.xticks(rotation=45)
plt.title('Dietary Fiber Content by Food Category')
plt.ylabel('Dietary Fiber (mg)')
plt.show()
```

### Scatter plot of Calories vs Total Fat

```python
# Scatter plot of Calories vs Total Fat
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Total Fat', y='Calories', data=data)
plt.title('Calories vs Total Fat')
plt.xlabel('Total Fat (g)')
plt.ylabel('Calories')
plt.show()
```

## Insights and Recommendations

### 1. Health-Conscious Menu Options:<br>
- Observation: 
    - Some items have extremely high calorie, fat, and sodium content (e.g., 1880 calories and 3600 mg sodium).<br>
- Recommendation:
    - Introduce more low-calorie, low-fat, and low-sodium options.
    - Highlight healthy choices on the menu with labels like "Heart-Healthy" or "Low Sodium."

### 2. Portion Control:
- Observation: 
    - Serving sizes and calorie counts vary significantly.
- Recommendation:
    - Offer smaller portion sizes for high-calorie items to cater to calorie-conscious customers.
    - Use clear labeling on serving sizes to help customers make informed decisions.

### 3. Focus on Dietary Fiber:
- Observation: 
    - The average dietary fiber content is low (1.63 g), with most items contributing less than the recommended daily intake.
- Recommendation:
    - Include more fiber-rich ingredients such as whole grains, vegetables, or legumes in the menu.
    - Add new high-fiber items like salads or grain bowls.

### 4. Reduce Sodium Levels:
- Observation: 
    - Some items exceed daily sodium intake recommendations (3600 mg max in the dataset).
- Recommendation: 
    - Reformulate recipes for high-sodium items to reduce sodium content.
    - Educate customers about sodium intake through menu notes or disclaimers.

### 5. Optimize High-Calorie Items:
- Observation: 
    - High-calorie items may attract customers seeking indulgence but can alienate health-focused diners.
- Recommendation:
    - Reintroduce high-calorie items in smaller sizes (e.g., "mini" or "snack" versions).
    -Pair them with healthier sides to balance the meal.

### 6. Enhance Customer Communication:
- Observation: 
    - Nutritional transparency is critical for building customer trust.
- Recommendation:
    - Create an interactive nutrition calculator on your website or app.
    - Provide clear nutritional breakdowns on menus to help customers make choices aligned with their dietary goals.

### 7. Target-Specific Audiences:
- Observation: 
    - Different menu categories (e.g., breakfast, desserts) serve diverse customer needs.
- Recommendation:
    - Promote breakfast items with a balance of protein and low calories for health-conscious morning diners.
    - Market indulgent desserts strategically for occasional treats.

### 8. Leverage Insights for Marketing:
- Observation: 
    - Popular categories or trends (e.g., low-sodium, high-protein) can drive sales.
- Recommendation:
    - Run marketing campaigns around healthier options like “Low-Calorie Month” or “High-Protein Specials.”
    - Use visuals (heatmaps, charts) in promotional materials to highlight improvements in health-conscious offerings.
